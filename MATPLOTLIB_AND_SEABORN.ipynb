```python
import seaborn as sns
import pandas as pd

# Creating Sample Data
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'IT', 'Finance', 'Sales', 'IT'],
    'Salary': [50000, 80000, 70000, 60000, 90000],
    'Experience': [3, 8, 5, 4, 10]
}

# Creating DataFrame
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Salary', y='Department', data=df, palette='coolwarm')

plt.title('Average Salary by Department')
plt.show()
```

```python
df
```

```python
sns.scatterplot(x='Experience', y='Salary', hue='Department', data=df, s = 250)

plt.title('Experience vs Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
sns.histplot(df['Salary'], bins=2, kde = True, color='green')

plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x=df['Salary'], color='orange')

plt.title('Salary Distribution (Box Plot)')
plt.show()
```
